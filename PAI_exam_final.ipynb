```python
%%writefile examples/locomotion/faulty_go2_wrapper.py
import torch
import numpy as np

class FaultyGo2Wrapper:
    """
    Go2Env をラップして脚ごとの '能力低下'(出力振幅縮小) イベントを挿入する。
    - 1エピソード中 各 env 4〜6 回 (デフォルト) ランダムステップで発火
    - 対象脚(3 DOF) の action 振幅を scale_range で乗算 (例 0.4〜0.6)
    - 観測: 既存 45 次元 + torque_mask(12) = 57 次元
    """
    def __init__(
        self,
        base_env,
        min_events=4,
        max_events=6,
        scale_range=(0.4, 0.6),
        seed=None,
        persistent=True,
        device=None
    ):
        self.env = base_env
        self.num_envs = base_env.num_envs
        self.num_actions = base_env.num_actions
        assert self.num_actions == 12, "Expect 12 actions for Go2"
        self.min_events = min_events
        self.max_events = max_events
        self.scale_range = scale_range
        self.persistent = persistent
        self.rng = np.random.default_rng(seed)
        self.device = device if device is not None else base_env.device

        self.joints_per_leg = 3
        self.num_legs = 4

        self.torque_mask = torch.ones((self.num_envs, self.num_actions), device=self.device)
        self.event_schedules = [[] for _ in range(self.num_envs)]
        self.legs_faulted = [set() for _ in range(self.num_envs)]

        # fault 発生瞬間検出用（extras に一時フラグ）
        self._last_fault_step = torch.full((self.num_envs,), -1, device=self.device, dtype=torch.long)

    # ---------- internal ----------
    def _sample_schedule(self):
        k = self.rng.integers(self.min_events, self.max_events + 1)
        max_len = self.env.max_episode_length
        candidates = np.arange(20, max_len - 20)
        if k > len(candidates):
            k = len(candidates)
        return sorted(self.rng.choice(candidates, size=k, replace=False).tolist())

    def _apply_fault_for_env(self, env_id):
        leg_id = self.rng.integers(0, self.num_legs)
        scale = self.rng.uniform(*self.scale_range)
        start = leg_id * self.joints_per_leg
        for j in range(self.joints_per_leg):
            self.torque_mask[env_id, start + j] = scale
        self.legs_faulted[env_id].add(leg_id)
        # step index は env.episode_length_buf に格納されるので後で参照
        self._last_fault_step[env_id] = self.env.episode_length_buf[env_id]

    def _maybe_inject_faults(self):
        steps = self.env.episode_length_buf  # tensor shape (num_envs,)
        for env_id in range(self.num_envs):
            sched = self.event_schedules[env_id]
            if not sched:
                continue
            if steps[env_id].item() == sched[0]:
                self._apply_fault_for_env(env_id)
                sched.pop(0)

    def _augment_obs(self, obs):
        return torch.cat([obs, self.torque_mask], dim=-1)

    def _reset_envs(self, envs_idx):
        if len(envs_idx) == 0:
            return
        for idx in envs_idx.tolist():
            self.event_schedules[idx] = self._sample_schedule()
            self.torque_mask[idx, :] = 1.0
            self.legs_faulted[idx].clear()
            self._last_fault_step[idx] = -1

    # ---------- public API ----------
    def reset(self):
        obs, _ = self.env.reset()
        for i in range(self.num_envs):
            self.event_schedules[i] = self._sample_schedule()
            self.torque_mask[i, :] = 1.0
            self.legs_faulted[i].clear()
            self._last_fault_step[i] = -1
        return self._augment_obs(obs), None

    def step(self, actions):
        scaled = actions * self.torque_mask
        obs, rew, reset, extras = self.env.step(scaled)
        # fault 挿入
        self._maybe_inject_faults()
        # リセットされた env 再初期化
        reset_envs = torch.nonzero(reset).reshape(-1)
        if len(reset_envs) > 0:
            self._reset_envs(reset_envs)

        obs_aug = self._augment_obs(obs)
        if extras is None:
            extras = {}
        extras = dict(extras)
        extras["torque_mask"] = self.torque_mask
        extras["torque_mask_mean"] = self.torque_mask.mean(dim=1)

        # fault event フラグ (1 ステップだけ 1.0)
        steps = self.env.episode_length_buf
        fault_event_flag = (steps == self._last_fault_step).float()
        extras["fault_event"] = fault_event_flag
        return obs_aug, rew, reset, extras

    @property
    def num_obs(self):
        return self.env.num_obs + self.num_actions  # 45 + 12 = 57

    @property
    def num_actions(self):
        return self.env.num_actions

    def get_observations(self):
        obs, extras = self.env.get_observations()
        return self._augment_obs(obs), extras

    def get_privileged_observations(self):
        # 現状 None
        return None
```

```python
%%writefile examples/locomotion/go2_env.py
import torch
import math
import genesis as gs
from genesis.utils.geom import transform_by_quat, inv_quat, transform_quat_by_quat

# ============================================================
# Quaternion -> Euler (roll, pitch, yaw) in degrees
# Assumes quaternion order (w, x, y, z).
# If挙動が変なら (x,y,z,w) だった可能性 → 関数内で並び替える。
# ============================================================
def quat_to_euler_wxyz(quat_tensor: torch.Tensor) -> torch.Tensor:
    """
    quat_tensor: (N,4) in (w,x,y,z)
    return: (N,3) degrees (roll, pitch, yaw)
    """
    w = quat_tensor[:, 0]
    x = quat_tensor[:, 1]
    y = quat_tensor[:, 2]
    z = quat_tensor[:, 3]

    # roll
    t0 = 2.0 * (w * x + y * z)
    t1 = 1.0 - 2.0 * (x * x + y * y)
    roll = torch.atan2(t0, t1)

    # pitch
    t2 = 2.0 * (w * y - z * x)
    t2 = torch.clamp(t2, -1.0, 1.0)
    pitch = torch.asin(t2)

    # yaw
    t3 = 2.0 * (w * z + x * y)
    t4 = 1.0 - 2.0 * (y * y + z * z)
    yaw = torch.atan2(t3, t4)

    return torch.stack([roll, pitch, yaw], dim=-1) * (180.0 / torch.pi)


def gs_rand_float(lower, upper, shape, device):
    return (upper - lower) * torch.rand(size=shape, device=device) + lower

class Go2Env:
    def __init__(self, num_envs, env_cfg, obs_cfg, reward_cfg, command_cfg, show_viewer=False):
        # ---- Device 決定（堅牢版） ----
        if not getattr(gs, "_initialized", False):
            try:
                gs.init(logging_level="warning")
            except Exception:
                pass

        def pick_device():
            # 1) GPU 優先
            if torch.cuda.is_available():
                try:
                    g = gs.get_gpu_device()
                    # 返り値がタプルの場合
                    if isinstance(g, (tuple, list)):
                        g = g[0]
                    # 文字列の場合
                    if isinstance(g, str):
                        return torch.device(g)
                    if isinstance(g, torch.device):
                        return g
                except Exception:
                    return torch.device("cuda:0")
            # 2) CPU
            try:
                c = gs.get_cpu_device()
                if isinstance(c, (tuple, list)):
                    c = c[0]
                if isinstance(c, str):
                    return torch.device(c)
                if isinstance(c, torch.device):
                    return c
            except Exception:
                pass
            return torch.device("cpu")

        self.device = pick_device()

        # ---- 既存コード続き ----
        self.num_envs = num_envs
        self.num_obs = obs_cfg["num_obs"]
        self.num_privileged_obs = None
        self.num_actions = env_cfg["num_actions"]
        self.num_commands = command_cfg["num_commands"]

        self.simulate_action_latency = True
        self.dt = 0.02
        self.max_episode_length = math.ceil(env_cfg["episode_length_s"] / self.dt)

        self.env_cfg = env_cfg
        self.obs_cfg = obs_cfg
        self.reward_cfg = reward_cfg
        self.command_cfg = command_cfg

        self.obs_scales = obs_cfg["obs_scales"]
        self.reward_scales = reward_cfg["reward_scales"]

        # ---- Scene（vis_options なし）----
        self.scene = gs.Scene(
            sim_options=gs.options.SimOptions(dt=self.dt, substeps=2),
            viewer_options=gs.options.ViewerOptions(
                max_FPS=int(0.5 / self.dt),
                camera_pos=(2.0, 0.0, 2.5),
                camera_lookat=(0.0, 0.0, 0.5),
                camera_fov=40,
            ),
            rigid_options=gs.options.RigidOptions(
                dt=self.dt,
                constraint_solver=gs.constraint_solver.Newton,
                enable_collision=True,
                enable_joint_limit=True,
            ),
            show_viewer=show_viewer,
        )

        # plane
        self.scene.add_entity(gs.morphs.URDF(file="urdf/plane/plane.urdf", fixed=True))

        # robot
        self.base_init_pos = torch.tensor(self.env_cfg["base_init_pos"], device=self.device)
        self.base_init_quat = torch.tensor(self.env_cfg["base_init_quat"], device=self.device)
        self.inv_base_init_quat = inv_quat(self.base_init_quat)
        self.robot = self.scene.add_entity(
            gs.morphs.URDF(
                file="urdf/go2/urdf/go2.urdf",
                pos=self.base_init_pos.cpu().numpy(),
                quat=self.base_init_quat.cpu().numpy(),
            ),
        )

        # build
        self.scene.build(n_envs=num_envs)

        # names → indices
        self.motors_dof_idx = [self.robot.get_joint(name).dof_start for name in self.env_cfg["joint_names"]]

        # PD
        self.robot.set_dofs_kp([self.env_cfg["kp"]] * self.num_actions, self.motors_dof_idx)
        self.robot.set_dofs_kv([self.env_cfg["kd"]] * self.num_actions, self.motors_dof_idx)

        # reward funcs
        self.reward_functions, self.episode_sums = dict(), dict()
        for name in self.reward_scales.keys():
            self.reward_scales[name] *= self.dt
            self.reward_functions[name] = getattr(self, "_reward_" + name)
            self.episode_sums[name] = torch.zeros((self.num_envs,), device=self.device, dtype=torch.float32)

        # buffers
        self.base_lin_vel = torch.zeros((self.num_envs, 3), device=self.device)
        self.base_ang_vel = torch.zeros((self.num_envs, 3), device=self.device)
        self.projected_gravity = torch.zeros((self.num_envs, 3), device=self.device)
        self.global_gravity = torch.tensor([0.0, 0.0, -1.0], device=self.device).repeat(self.num_envs, 1)
        self.obs_buf = torch.zeros((self.num_envs, self.num_obs), device=self.device)
        self.rew_buf = torch.zeros((self.num_envs,), device=self.device)
        self.reset_buf = torch.ones((self.num_envs,), device=self.device, dtype=torch.int32)
        self.episode_length_buf = torch.zeros((self.num_envs,), device=self.device, dtype=torch.int32)
        self.commands = torch.zeros((self.num_envs, self.num_commands), device=self.device)
        self.commands_scale = torch.tensor(
            [self.obs_scales["lin_vel"], self.obs_scales["lin_vel"], self.obs_scales["ang_vel"]],
            device=self.device,
        )
        self.actions = torch.zeros((self.num_envs, self.num_actions), device=self.device)
        self.last_actions = torch.zeros_like(self.actions)
        self.dof_pos = torch.zeros_like(self.actions)
        self.dof_vel = torch.zeros_like(self.actions)
        self.last_dof_vel = torch.zeros_like(self.actions)
        self.base_pos = torch.zeros((self.num_envs, 3), device=self.device)
        self.base_quat = torch.zeros((self.num_envs, 4), device=self.device)
        self.default_dof_pos = torch.tensor(
            [self.env_cfg["default_joint_angles"][n] for n in self.env_cfg["joint_names"]],
            device=self.device,
        )
        self.extras = {"observations": {}}

    # -------------------------------------------------
    def _resample_commands(self, envs_idx):
        self.commands[envs_idx, 0] = gs_rand_float(*self.command_cfg["lin_vel_x_range"], (len(envs_idx),), gs.device)
        self.commands[envs_idx, 1] = gs_rand_float(*self.command_cfg["lin_vel_y_range"], (len(envs_idx),), gs.device)
        self.commands[envs_idx, 2] = gs_rand_float(*self.command_cfg["ang_vel_range"], (len(envs_idx),), gs.device)

    # -------------------------------------------------
    def step(self, actions):
        self.actions = torch.clip(actions, -self.env_cfg["clip_actions"], self.env_cfg["clip_actions"])
        exec_actions = self.last_actions if self.simulate_action_latency else self.actions
        target_dof_pos = exec_actions * self.env_cfg["action_scale"] + self.default_dof_pos
        self.robot.control_dofs_position(target_dof_pos, self.motors_dof_idx)
        self.scene.step()

        # update state
        self.episode_length_buf += 1
        self.base_pos[:] = self.robot.get_pos()
        self.base_quat[:] = self.robot.get_quat()

        rel_quat = transform_quat_by_quat(torch.ones_like(self.base_quat) * self.inv_base_init_quat, self.base_quat)
        self.base_euler = quat_to_euler_wxyz(rel_quat)  # (N,3) degrees: roll, pitch, yaw

        inv_base_quat = inv_quat(self.base_quat)
        self.base_lin_vel[:] = transform_by_quat(self.robot.get_vel(), inv_base_quat)
        self.base_ang_vel[:] = transform_by_quat(self.robot.get_ang(), inv_base_quat)
        self.projected_gravity = transform_by_quat(self.global_gravity, inv_base_quat)
        self.dof_pos[:] = self.robot.get_dofs_position(self.motors_dof_idx)
        self.dof_vel[:] = self.robot.get_dofs_velocity(self.motors_dof_idx)

        # resample commands
        envs_idx = (
            (self.episode_length_buf % int(self.env_cfg["resampling_time_s"] / self.dt) == 0)
            .nonzero(as_tuple=False)
            .reshape((-1,))
        )
        self._resample_commands(envs_idx)

        # terminations
        self.reset_buf = self.episode_length_buf > self.max_episode_length
        self.reset_buf |= torch.abs(self.base_euler[:, 1]) > self.env_cfg["termination_if_pitch_greater_than"]
        self.reset_buf |= torch.abs(self.base_euler[:, 0]) > self.env_cfg["termination_if_roll_greater_than"]

        time_out_idx = (self.episode_length_buf > self.max_episode_length).nonzero(as_tuple=False).reshape((-1,))
        self.extras["time_outs"] = torch.zeros_like(self.reset_buf, device=gs.device, dtype=gs.tc_float)
        self.extras["time_outs"][time_out_idx] = 1.0

        self.reset_idx(self.reset_buf.nonzero(as_tuple=False).reshape((-1,)))

        # reward
        self.rew_buf[:] = 0.0
        for name, reward_func in self.reward_functions.items():
            rew = reward_func() * self.reward_scales[name]
            self.rew_buf += rew
            self.episode_sums[name] += rew

        # observations (45)
        self.obs_buf = torch.cat(
            [
                self.base_ang_vel * self.obs_scales["ang_vel"],            # 3
                self.projected_gravity,                                    # 3
                self.commands * self.commands_scale,                       # 3
                (self.dof_pos - self.default_dof_pos) * self.obs_scales["dof_pos"],  # 12
                self.dof_vel * self.obs_scales["dof_vel"],                 # 12
                self.actions,                                              # 12
            ],
            axis=-1,
        )

        self.last_actions[:] = self.actions[:]
        self.last_dof_vel[:] = self.dof_vel[:]

        self.extras["observations"]["critic"] = self.obs_buf

        return self.obs_buf, self.rew_buf, self.reset_buf, self.extras

    # -------------------------------------------------
    def get_observations(self):
        self.extras["observations"]["critic"] = self.obs_buf
        return self.obs_buf, self.extras

    def get_privileged_observations(self):
        return None

    # -------------------------------------------------
    def reset_idx(self, envs_idx):
        if len(envs_idx) == 0:
            return
        self.dof_pos[envs_idx] = self.default_dof_pos
        self.dof_vel[envs_idx] = 0.0
        self.robot.set_dofs_position(
            position=self.dof_pos[envs_idx],
            dofs_idx_local=self.motors_dof_idx,
            zero_velocity=True,
            envs_idx=envs_idx,
        )
        self.base_pos[envs_idx] = self.base_init_pos
        self.base_quat[envs_idx] = self.base_init_quat.reshape(1, -1)
        self.robot.set_pos(self.base_pos[envs_idx], zero_velocity=False, envs_idx=envs_idx)
        self.robot.set_quat(self.base_quat[envs_idx], zero_velocity=False, envs_idx=envs_idx)
        self.base_lin_vel[envs_idx] = 0
        self.base_ang_vel[envs_idx] = 0
        self.robot.zero_all_dofs_velocity(envs_idx)

        self.last_actions[envs_idx] = 0.0
        self.last_dof_vel[envs_idx] = 0.0
        self.episode_length_buf[envs_idx] = 0
        self.reset_buf[envs_idx] = True

        self.extras["episode"] = {}
        for key in self.episode_sums.keys():
            self.extras["episode"]["rew_" + key] = (
                torch.mean(self.episode_sums[key][envs_idx]).item() / self.env_cfg["episode_length_s"]
            )
            self.episode_sums[key][envs_idx] = 0.0

        self._resample_commands(envs_idx)

    # -------------------------------------------------
    def reset(self):
        self.reset_buf[:] = True
        self.reset_idx(torch.arange(self.num_envs, device=gs.device))
        return self.obs_buf, None

    # ---------------- Reward functions ----------------
    def _reward_tracking_lin_vel(self):
        lin_vel_error = torch.sum(torch.square(self.commands[:, :2] - self.base_lin_vel[:, :2]), dim=1)
        return torch.exp(-lin_vel_error / self.reward_cfg["tracking_sigma"])

    def _reward_tracking_ang_vel(self):
        ang_vel_error = torch.square(self.commands[:, 2] - self.base_ang_vel[:, 2])
        return torch.exp(-ang_vel_error / self.reward_cfg["tracking_sigma"])

    def _reward_lin_vel_z(self):
        return torch.square(self.base_lin_vel[:, 2])

    def _reward_action_rate(self):
        return torch.sum(torch.square(self.last_actions - self.actions), dim=1)

    def _reward_similar_to_default(self):
        return torch.sum(torch.abs(self.dof_pos - self.default_dof_pos), dim=1)

    def _reward_base_height(self):
        return torch.square(self.base_pos[:, 2] - self.reward_cfg["base_height_target"])
```

```python
%%writefile examples/locomotion/faulty_go2_wrapper.py
import torch
import numpy as np

class FaultyGo2Wrapper:
    """
    Go2Env をラップして脚ごとの '能力低下'(出力振幅縮小) イベントを挿入する。
    観測: 元 45 + torque_mask(12) = 57
    """
    def __init__(
        self,
        base_env,
        min_events=4,
        max_events=6,
        scale_range=(0.4, 0.6),
        seed=None,
        persistent=True,
        device=None
    ):
        self.env = base_env
        self.num_envs = base_env.num_envs
        # num_actions はプロパティで参照するので代入しない
        assert base_env.num_actions == 12, "Expect 12 actions for Go2"
        self.min_events = min_events
        self.max_events = max_events
        self.scale_range = scale_range
        self.persistent = persistent
        self.rng = np.random.default_rng(seed)
        self.device = device if device is not None else base_env.device

        self.joints_per_leg = 3
        self.num_legs = 4

        self.torque_mask = torch.ones(
            (self.num_envs, self.env.num_actions),
            device=self.device
        )
        self.event_schedules = [[] for _ in range(self.num_envs)]
        self.legs_faulted = [set() for _ in range(self.num_envs)]
        self._last_fault_step = torch.full((self.num_envs,), -1, device=self.device, dtype=torch.long)

    # ---------- internal ----------
    def _sample_schedule(self):
        k = self.rng.integers(self.min_events, self.max_events + 1)
        max_len = self.env.max_episode_length
        candidates = np.arange(20, max_len - 20)
        if k > len(candidates):
            k = len(candidates)
        return sorted(self.rng.choice(candidates, size=k, replace=False).tolist())

    def _apply_fault_for_env(self, env_id):
        leg_id = self.rng.integers(0, self.num_legs)
        scale = self.rng.uniform(*self.scale_range)
        start = leg_id * self.joints_per_leg
        for j in range(self.joints_per_leg):
            self.torque_mask[env_id, start + j] = scale
        self.legs_faulted[env_id].add(leg_id)
        self._last_fault_step[env_id] = self.env.episode_length_buf[env_id]

    def _maybe_inject_faults(self):
        steps = self.env.episode_length_buf
        for env_id in range(self.num_envs):
            sched = self.event_schedules[env_id]
            if not sched:
                continue
            if steps[env_id].item() == sched[0]:
                self._apply_fault_for_env(env_id)
                sched.pop(0)

    def _augment_obs(self, obs):
        return torch.cat([obs, self.torque_mask], dim=-1)

    def _reset_envs(self, envs_idx):
        if len(envs_idx) == 0:
            return
        for idx in envs_idx.tolist():
            self.event_schedules[idx] = self._sample_schedule()
            self.torque_mask[idx, :] = 1.0
            self.legs_faulted[idx].clear()
            self._last_fault_step[idx] = -1

    # ---------- public API ----------
    def reset(self):
        obs, _ = self.env.reset()
        for i in range(self.num_envs):
            self.event_schedules[i] = self._sample_schedule()
            self.torque_mask[i, :] = 1.0
            self.legs_faulted[i].clear()
            self._last_fault_step[i] = -1
        return self._augment_obs(obs), None

    def step(self, actions):
        scaled = actions * self.torque_mask
        obs, rew, reset, extras = self.env.step(scaled)
        self._maybe_inject_faults()
        reset_envs = torch.nonzero(reset).reshape(-1)
        if len(reset_envs) > 0:
            self._reset_envs(reset_envs)

        obs_aug = self._augment_obs(obs)
        if extras is None:
            extras = {}
        extras = dict(extras)
        extras["torque_mask"] = self.torque_mask
        extras["torque_mask_mean"] = self.torque_mask.mean(dim=1)

        steps = self.env.episode_length_buf
        fault_event_flag = (steps == self._last_fault_step).float()
        extras["fault_event"] = fault_event_flag
        return obs_aug, rew, reset, extras

    @property
    def num_obs(self):
        return self.env.num_obs + self.env.num_actions  # 45 + 12 = 57

    @property
    def num_actions(self):
        return self.env.num_actions

    def get_observations(self):
        obs, extras = self.env.get_observations()
        return self._augment_obs(obs), extras

    def get_privileged_observations(self):
        return None
```

```python
%%writefile examples/locomotion/faulty_go2_wrapper.py
import torch, numpy as np, math

class FaultyGo2Wrapper:
    """
    Go2 ベース環境を耐故障拡張するラッパ
    - 故障モデル: 選ばれた脚のトルク（=アクション) を mask でスケールダウン
      mask_leg(t) = 1 - (1-s0)*exp(-k*t)  (s0: 初期スケール 0.4~0.6 など)
    - 発生方式:
        (A) スケジュール: 各 env に故障発生ステップを事前サンプリング
        (B) 確率: 各ステップ p で新規故障（最大同時数制限あり）
      → `prob_new_fault` が None ならスケジュール / float >0 なら確率方式
    - 観測: base_obs(45) + torque_mask(12) = 57
    - ログ:
        extras に以下を追加
          torque_mask, fault_event (envごと bool),
          fault_active_mask (N,4), fault_scale_mask (N,4),
          newly_recovered_mask, num_active_faults
    - エピソード統計（reset時に extras["episode"] へ）
          mean_fault_duration, faults_count
    """

    def __init__(
        self,
        base_env,
        min_events=2,
        max_events=3,
        scale_range=(0.4,0.6),
        recovery_k=2.0,
        prob_new_fault=None,          # None=スケジュール / float>0=確率
        max_concurrent_faults=2,
        seed=0,
        recovery_done_threshold=0.995,
        device=None
    ):
        self.env = base_env
        self.min_events = min_events
        self.max_events = max_events
        self.scale_range = scale_range
        self.recovery_k = recovery_k
        self.prob_new_fault = prob_new_fault
        self.max_concurrent_faults = max_concurrent_faults
        self.recovery_done_th = recovery_done_threshold

        self.device = device or torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.rng = np.random.default_rng(seed)

        self.joints_per_leg = 3
        self.num_legs = 4
        N = self.env.num_envs
        A = self.env.num_actions

        # マスク・状態
        self.torque_mask = torch.ones((N, A), device=self.device)
        self.fault_active = torch.zeros((N, self.num_legs), dtype=torch.bool, device=self.device)
        self.fault_scale_init = torch.ones((N, self.num_legs), device=self.device)
        self.fault_time = torch.zeros((N, self.num_legs), device=self.device)

        self.last_fault_step = torch.full((N,), -1, dtype=torch.long, device=self.device)
        self.newly_recovered = torch.zeros((N, self.num_legs), dtype=torch.bool, device=self.device)

        # durations
        self._active_start_step = torch.full((N, self.num_legs), -1, dtype=torch.long, device=self.device)
        self._fault_durations_accum = [[] for _ in range(N)]  # list of durations per env

        # スケジュール（prob_new_fault=None のとき使用）
        self.event_schedules = [[] for _ in range(N)]
        if self.prob_new_fault is None:
            self._resample_all_schedules()

        # 観測サイズ
        self.num_obs = self.env.num_obs + A
        self.num_privileged_obs = None  # RSL-RL 互換用

    # ---- properties 透過 (Runner互換) ----
    @property
    def episode_length_buf(self): return self.env.episode_length_buf
    @episode_length_buf.setter
    def episode_length_buf(self, v): self.env.episode_length_buf = v
    @property
    def max_episode_length(self): return self.env.max_episode_length
    @property
    def dt(self): return self.env.dt
    @property
    def num_envs(self): return self.env.num_envs
    @property
    def num_actions(self): return self.env.num_actions

    # ---- スケジュール生成 ----
    def _sample_schedule(self):
        k = self.rng.integers(self.min_events, self.max_events+1)
        max_len = self.env.max_episode_length
        # 端を避ける
        candidates = np.arange(20, max(21, max_len-20))
        if k > len(candidates):
            k = len(candidates)
        if k <= 0:
            return []
        steps = self.rng.choice(candidates, size=k, replace=False)
        return sorted(steps.tolist())

    def _resample_all_schedules(self, envs_idx=None):
        if envs_idx is None:
            envs_idx = range(self.env.num_envs)
        for i in envs_idx:
            self.event_schedules[i] = self._sample_schedule()

    # ---- 故障発生 ----
    def _activate_fault(self, env_id):
        # 空き脚
        active = self.fault_active[env_id]
        if active.sum() >= self.max_concurrent_faults:
            return
        # まだ故障でない脚
        candidates = [l for l in range(self.num_legs) if not active[l]]
        if not candidates:
            return
        leg = self.rng.integers(0, len(candidates))
        leg_id = candidates[leg]
        scale0 = self.rng.uniform(*self.scale_range)
        self.fault_active[env_id, leg_id] = True
        self.fault_scale_init[env_id, leg_id] = scale0
        self.fault_time[env_id, leg_id] = 0.0
        self.last_fault_step[env_id] = self.episode_length_buf[env_id].item()
        self._active_start_step[env_id, leg_id] = self.episode_length_buf[env_id].item()

    # ---- マスク更新 ----
    def _update_masks(self):
        self.torque_mask[:] = 1.0
        # 時間更新
        self.fault_time[self.fault_active] += self.dt
        # 再計算
        active_idx = torch.nonzero(self.fault_active, as_tuple=False)
        for env_id, leg_id in active_idx.tolist():
            s0 = self.fault_scale_init[env_id, leg_id]
            t = self.fault_time[env_id, leg_id]
            mask_leg = 1.0 - (1.0 - s0) * math.exp(-self.recovery_k * float(t))
            start = leg_id * self.joints_per_leg
            self.torque_mask[env_id, start:start+self.joints_per_leg] = mask_leg
        # 回復判定
        self.newly_recovered[:] = False
        for env_id, leg_id in active_idx.tolist():
            start = leg_id * self.joints_per_leg
            mval = float(self.torque_mask[env_id, start])
            if mval >= self.recovery_done_th:
                self.newly_recovered[env_id, leg_id] = True
                self.fault_active[env_id, leg_id] = False
                # duration ログ
                st = self._active_start_step[env_id, leg_id].item()
                if st >= 0:
                    dur = self.episode_length_buf[env_id].item() - st
                    self._fault_durations_accum[env_id].append(dur * self.dt)
                self._active_start_step[env_id, leg_id] = -1

    # ---- 故障トリガ ----
    def _maybe_trigger(self):
        steps = self.episode_length_buf
        if self.prob_new_fault is None:
            # スケジュール方式
            for e in range(self.env.num_envs):
                sched = self.event_schedules[e]
                if sched and steps[e].item() == sched[0]:
                    self._activate_fault(e)
                    sched.pop(0)
        else:
            # 確率方式
            for e in range(self.env.num_envs):
                if np.random.rand() < self.prob_new_fault:
                    self._activate_fault(e)

    # ---- 観測拡張 ----
    def _augment_obs(self, obs):
        return torch.cat([obs, self.torque_mask], dim=-1)

    # ---- リセット ----
    def reset(self):
        obs, _ = self.env.reset()
        self.torque_mask[:] = 1.0
        self.fault_active[:] = False
        self.fault_scale_init[:] = 1.0
        self.fault_time[:] = 0.0
        self.last_fault_step[:] = -1
        self.newly_recovered[:] = False
        self._active_start_step[:] = -1
        self._fault_durations_accum = [[] for _ in range(self.env.num_envs)]
        if self.prob_new_fault is None:
            self._resample_all_schedules()
        return self._augment_obs(obs), {}

    # ---- ステップ ----
    def step(self, actions):
        # 故障更新 & スケジュール
        self._maybe_trigger()
        self._update_masks()

        scaled = actions * self.torque_mask
        obs, rew, reset, extras = self.env.step(scaled)
        if extras is None:
            extras = {}
        else:
            extras = dict(extras)

        # ログメタ
        steps = self.episode_length_buf
        fault_event = (steps == self.last_fault_step)

        extras["torque_mask"] = self.torque_mask
        extras["fault_event"] = fault_event
        extras["fault_active_mask"] = self.fault_active
        # 脚ごとの現在マスク平均（各脚3関節→代表値として最初のジョイント値）
        leg_vals = []
        for leg in range(self.num_legs):
            leg_vals.append(self.torque_mask[:, leg*self.joints_per_leg].unsqueeze(1))
        extras["fault_scale_mask"] = torch.cat(leg_vals, dim=1)
        extras["newly_recovered_mask"] = self.newly_recovered
        extras["num_active_faults"] = self.fault_active.sum(dim=1)

        # リセット時にエピソード統計
        reset_envs = torch.nonzero(reset).reshape(-1)
        if len(reset_envs) > 0:
            if "episode" not in extras or extras["episode"] is None:
                extras["episode"] = {}
            for e in reset_envs.tolist():
                ds = self._fault_durations_accum[e]
                mean_d = float(np.mean(ds)) if ds else 0.0
                extras["episode"]["mean_fault_duration"] = mean_d
                extras["episode"]["faults_count"] = len(ds)

        return self._augment_obs(obs), rew, reset, extras

    def get_observations(self):
        obs, ex = self.env.get_observations()
        return self._augment_obs(obs), ex

    def get_privileged_observations(self):
        return None
```

```python
import torch, math, numpy as np, json, os

# メトリク設定
VEL_ERR_RECOVERY_THRESH = 0.10   # (m/s) 速度誤差がこれ未満なら回復達成条件の一部
ROLL_PITCH_RECOVERY_DEG = 6.0    # 姿勢安定条件
RECOVERY_WINDOW_MAX = 5.0        # 故障後この秒数までに回復判定を探す
PEAK_WINDOW = 0.6                # 故障直後ショックのピークを計測する秒数
INTEGRAL_WINDOW = 3.0            # 故障後 誤差積分窓

def quaternion_to_euler_wxyz(q):
    # q: (w,x,y,z)
    w,x,y,z = q
    # roll (x), pitch (y), yaw (z)
    t0 = +2.0*(w*x + y*z)
    t1 = +1.0 - 2.0*(x*x + y*y)
    roll = math.atan2(t0, t1)

    t2 = +2.0*(w*y - z*x)
    t2 = max(-1.0, min(1.0, t2))
    pitch = math.asin(t2)

    t3 = +2.0*(w*z + x*y)
    t4 = +1.0 - 2.0*(y*y + z*z)
    yaw = math.atan2(t3, t4)
    return roll, pitch, yaw  # (rad)


def compute_recovery_metrics(log):
    """
    log: fault_run_raw.json をロードした dict
    Returns:
        metrics: 各故障イベントの辞書リスト
        summary: 集計
    """
    t = np.array(log["t"])
    v_cmd = np.array(log["cmd_vx"])
    v_base = np.array(log["base_vx"])
    roll = np.array(log["roll"])
    pitch = np.array(log["pitch"])
    dt = log["dt"]

    fault_steps = log.get("fault_steps", [])
    fault_legs  = log.get("fault_leg_at_event", [])
    mask_mean_at_fault = log.get("mask_mean_at_fault", [])

    # 安全: 長さ不整合を最小長に揃える
    n_faults = min(len(fault_steps), len(fault_legs), len(mask_mean_at_fault))

    v_err = np.abs(v_cmd - v_base)
    roll_deg_full = np.abs(roll * 180/np.pi)
    pitch_deg_full = np.abs(pitch * 180/np.pi)

    metrics = []
    for i_fault in range(n_faults):
        fs = fault_steps[i_fault]          # ステップ index
        leg = fault_legs[i_fault]
        mask0 = mask_mean_at_fault[i_fault]
        if fs >= len(t):
            # 念のため (早期終了などでステップが欠けたケース)
            continue

        t_fault = t[fs]

        # 回復条件
        post_mask = (t >= t_fault)
        recover_condition = post_mask & \
            (v_err < VEL_ERR_RECOVERY_THRESH) & \
            (roll_deg_full < ROLL_PITCH_RECOVERY_DEG) & \
            (pitch_deg_full < ROLL_PITCH_RECOVERY_DEG)

        recover_time = None
        # 故障直後(同ステップ)は除外
        for idx in np.where(recover_condition)[0]:
            if t[idx] > t_fault:
                recover_time = t[idx] - t_fault
                break

        # Peak error
        peak_mask = (t >= t_fault) & (t <= t_fault + PEAK_WINDOW)
        peak_err = float(v_err[peak_mask].max()) if peak_mask.any() else float("nan")

        # Integrated error
        integ_mask = (t >= t_fault) & (t <= t_fault + INTEGRAL_WINDOW)
        integral_err = float(np.trapz(v_err[integ_mask], t[integ_mask])) if integ_mask.any() else float("nan")

        metrics.append({
            "fault_time": float(t_fault),
            "recover_time": float(recover_time) if recover_time is not None else None,
            "peak_vel_err": peak_err,
            "integrated_vel_err": integral_err,
            "fault_leg": int(leg),
            "mask_initial_mean": float(mask0),
        })

    # サマリ
    valid_recover = [m["recover_time"] for m in metrics if m["recover_time"] is not None]
    summary = {
        "num_faults": len(metrics),
        "mean_recover_time": float(np.mean(valid_recover)) if valid_recover else None,
        "mean_peak_vel_err": float(np.mean([m["peak_vel_err"] for m in metrics])) if metrics else None,
        "mean_integrated_vel_err": float(np.mean([m["integrated_vel_err"] for m in metrics])) if metrics else None,
    }
    return metrics, summary
```

```python
import json, numpy as np, imageio, os, math
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from matplotlib import cm

with open("logs/fault_run_3d.json","r") as f:
    L = json.load(f)

t = np.array(L["t"])
base_pos = np.array(L["base_pos"])       # (N,3)
base_quat = np.array(L["base_quat"])     # (N,4)
fault_leg_ids = np.array(L["fault_leg_ids"])
fault_active_mask = np.array(L["fault_active_mask"])
torque_mask = np.array(L["torque_mask"]) # (N,12)
fault_event = np.array(L["fault_event"])
dt = L["dt"]

# 脚オフセット（単純）
L_body = 0.38
W_body = 0.21
h_foot = -0.22  # ベース中心からフット (見た目調整)
foot_offsets_local = {
    0: np.array([ L_body/2, -W_body/2, h_foot]),
    1: np.array([ L_body/2,  W_body/2, h_foot]),
    2: np.array([-L_body/2, -W_body/2, h_foot]),
    3: np.array([-L_body/2,  W_body/2, h_foot]),
}

# 四元数 → 回転行列 (w,x,y,z) 仮定
def quat_to_R(q):
    # ヘuristic: 大きい成分が w か末尾か
    if abs(q[0]) < 0.5 and abs(q[3]) > 0.5:
        w,x,y,z = q[3], q[0], q[1], q[2]
    else:
        w,x,y,z = q
    return np.array([
        [1-2*(y*y+z*z), 2*(x*y - z*w),   2*(x*z + y*w)],
        [2*(x*y + z*w),   1-2*(x*x+z*z), 2*(y*z - x*w)],
        [2*(x*z - y*w),   2*(y*z + x*w), 1-2*(x*x+y*y)]
    ])

# カラーマップ: torque mask の “脚単位平均” へ適用
cmap = cm.get_cmap("plasma")
base_leg_colors = {
    0: "#1f77b4",
    1: "#ff7f0e",
    2: "#2ca02c",
    3: "#d62728"
}

def mask_color(v):
    n=(v-0.4)/0.6
    n=np.clip(n,0,1)
    r,g,b,a=cmap(n)
    return (r,g,b)

frames=[]
# フレーム間引き: 2ステップに1枚 → ほぼ 25fps で 60秒
indices = list(range(0, len(t), 2))

for k, idx in enumerate(indices):
    bp = base_pos[idx]
    bq = base_quat[idx]

    Rm = quat_to_R(bq)
    leg_colors_draw = {}

    # 図設定
    fig = plt.figure(figsize=(5.2,5.2))
    ax = fig.add_subplot(111, projection='3d')
    ax.set_facecolor("white")
    # カメラ中心を base 周りに
    R_lim = 0.7
    ax.set_xlim(bp[0]-R_lim, bp[0]+R_lim)
    ax.set_ylim(bp[1]-R_lim, bp[1]+R_lim)
    ax.set_zlim(bp[2]-0.35, bp[2]+0.40)
    ax.set_xticks([]); ax.set_yticks([]); ax.set_zticks([])

    # ボディ (線で簡易枠)
    # 角点 (ローカル)
    corners = np.array([
        [ L_body/2,  W_body/2, 0.0],
        [ L_body/2, -W_body/2, 0.0],
        [-L_body/2, -W_body/2, 0.0],
        [-L_body/2,  W_body/2, 0.0],
    ])
    body_world = (Rm @ corners.T).T + bp
    # 閉ループ描画
    loop = np.vstack([body_world, body_world[0]])
    ax.plot(loop[:,0], loop[:,1], loop[:,2], c="#444444", linewidth=2)

    # Fault 状態判定
    new_fault = (fault_event[idx] > 0.5)
    current_fault_leg = fault_leg_ids[idx]
    active_mask_row = fault_active_mask[idx]  # shape (4,)

    # 各脚
    for leg_id in range(4):
        foot_local = foot_offsets_local[leg_id]
        foot_world = bp + Rm @ foot_local
        # mask の該当3関節平均
        joint_slice = torque_mask[idx][leg_id*3:(leg_id+1)*3]
        mean_leg_mask = float(np.mean(joint_slice))

        if new_fault and leg_id == current_fault_leg:
            color = "#FFD700"
        elif active_mask_row[leg_id] == 1:
            # 回復中
            color = mask_color(mean_leg_mask)
        else:
            color = base_leg_colors[leg_id]

        leg_colors_draw[leg_id]=color
        # 脚ライン
        ax.plot([bp[0], foot_world[0]], [bp[1], foot_world[1]], [bp[2], foot_world[2]],
                c=color, linewidth=3, alpha=0.95)
        # 足先
        ax.scatter([foot_world[0]],[foot_world[1]],[foot_world[2]], c=[color], s=40)

    # タイトル / HUD
    time_txt = f"t={t[idx]:.1f}s"
    if new_fault:
        time_txt += f"  FAULT: Leg {current_fault_leg}"
    ax.set_title(time_txt, fontsize=11, color=("red" if new_fault else "black"))

    # 凡例簡略（初回か fault 毎）
    if new_fault or k < 3:
        ax.text2D(0.02,0.02,
                  "Gold: New Fault\nColor gradient: Recovery\nBase colors: Healthy",
                  transform=ax.transAxes, fontsize=8,
                  bbox=dict(facecolor='white', alpha=0.7, boxstyle='round'))

    fig.canvas.draw()
    buf = np.frombuffer(fig.canvas.buffer_rgba(), dtype=np.uint8)
    w,h = fig.canvas.get_width_height()
    img = buf.reshape(h,w,4)[...,:3]
    frames.append(img)
    plt.close(fig)

# FPS 調整 → 60秒狙い
duration_sec = t[indices[-1]] - t[indices[0]]
fps = max(10, int(len(frames)/duration_sec))
os.makedirs("videos", exist_ok=True)
out_path = "videos/fault_story_3d_simple.mp4"
imageio.mimsave(out_path, frames, fps=fps)
print(f"Saved {out_path}  frames={len(frames)}  fps={fps}  ≈{len(frames)/fps:.1f}s")
```

```python
%%writefile task_eval.py
import argparse, json, math, torch, numpy as np
import genesis as gs

# Go2 環境（公式 examples へのパス想定）
import sys, os
sys.path.append("examples/locomotion")
from go2_env import Go2Env
from go2_train import get_cfgs

from faulty_go2_wrapper import FaultyGo2Wrapper

def parse_waypoints(s):
    # 例: "(0,0),(2.0,0),(3.5,0.8)"
    s = s.strip()
    pts=[]
    for part in s.split(")"):
        part=part.strip().lstrip(",").strip()
        if not part: continue
        part = part.lstrip("(").rstrip(",")
        if not part: continue
        xy = part.split(",")
        if len(xy)!=2: continue
        pts.append((float(xy[0]), float(xy[1])))
    return pts

def main():
    ap = argparse.ArgumentParser()
    ap.add_argument("--episode-sec", type=float, default=60.0)
    ap.add_argument("--waypoints", type=str, default="(0,0),(2,0),(3.5,0.8),(5,0.8)")
    ap.add_argument("--wp-radius", type=float, default=0.35)
    ap.add_argument("--push-times", type=str, default="28.0")
    ap.add_argument("--push-force", type=str, default="0,150,0")  # world frame impulse-like（粗い実装）
    ap.add_argument("--num-envs", type=int, default=1)
    ap.add_argument("--prob-new-fault", type=float, default=0.04)
    ap.add_argument("--min-events", type=int, default=2)
    ap.add_argument("--max-events", type=int, default=3)
    ap.add_argument("--scale-range", type=str, default="0.4,0.6")
    ap.add_argument("--recovery-k", type=float, default=2.0)
    ap.add_argument("--outfile", type=str, default="task_log.json")
    ap.add_argument("--policy-random-std", type=float, default=0.15, help="ランダム行動の標準偏差")
    ap.add_argument("--seed", type=int, default=123)
    args = ap.parse_args()

    gs.init(logging_level="warning")

    env_cfg, obs_cfg, reward_cfg, command_cfg = get_cfgs()
    env_cfg["episode_length_s"] = args.episode_sec  # エピソード長合わせ

    # ベース環境
    base = Go2Env(
        num_envs=args.num_envs,
        env_cfg=env_cfg,
        obs_cfg=obs_cfg,
        reward_cfg=reward_cfg,
        command_cfg=command_cfg,
        show_viewer=False
    )

    scale_lo, scale_hi = [float(x) for x in args.scale_range.split(",")]
    wrapper = FaultyGo2Wrapper(
        base_env=base,
        min_events=args.min_events,
        max_events=args.max_events,
        scale_range=(scale_lo, scale_hi),
        recovery_k=args.recovery_k,
        prob_new_fault=args.prob_new_fault,
        seed=args.seed
    )

    waypoints = parse_waypoints(args.waypoints)
    push_times = [float(x) for x in args.push_times.split(",") if x.strip()]
    push_force = tuple(float(x) for x in args.push_force.split(","))

    obs,_ = wrapper.reset()
    dt = base.dt
    total_steps = int(args.episode_sec / dt)

    log = dict(
        t=[],
        base_pos=[],
        base_lin_vel=[],
        torque_mask_mean=[],
        fault_event=[],
        active_faults=[],
        wp_index=[],
        push_flag=[],
        leg_mask=[],      # 脚単位マスク (N=1想定) shape steps x 4
        leg_fault_active=[],
    )

    cur_wp = 0
    torch.manual_seed(args.seed)
    np.random.seed(args.seed)

    def apply_push_conditionally(step_time):
        # 粗い外乱：ベース速度に上乗せ（シンプル）
        if any(abs(step_time - pt) < 0.5*dt for pt in push_times):
            v = base.base_lin_vel
            # add impulse-like velocity (world frame y)
            v[:,0] += 0.0
            v[:,1] += push_force[1] / 1000.0  # スケール調整
            base.base_lin_vel[:] = v
            return True
        return False

    for step in range(total_steps):
        # ---- 行動（ここに学習済みポリシーを差し替え可能）----
        actions = torch.randn((base.num_envs, base.num_actions)) * args.policy_random_std

        obs, rew, reset, extras = wrapper.step(actions)

        t = step * dt
        base_pos = base.base_pos[0].detach().cpu().numpy()
        base_lin_vel = base.base_lin_vel[0].detach().cpu().numpy()

        # Waypoint 判定
        if cur_wp < len(waypoints):
            wx, wy = waypoints[cur_wp]
            if np.linalg.norm(base_pos[:2] - np.array([wx, wy])) < args.wp_radius:
                cur_wp += 1

        pushed = apply_push_conditionally(t)

        # ロギング
        log["t"].append(t)
        log["base_pos"].append(base_pos.tolist())
        log["base_lin_vel"].append(base_lin_vel.tolist())
        log["torque_mask_mean"].append(float(extras["torque_mask"][0].mean().item()))
        log["fault_event"].append(int(extras["fault_event"][0].item()))
        log["active_faults"].append(int(extras["num_active_faults"][0].item()))
        log["wp_index"].append(cur_wp)
        log["push_flag"].append(int(pushed))

        leg_mask_vals = extras["fault_scale_mask"][0].detach().cpu().numpy().tolist()
        leg_act_vals = extras["fault_active_mask"][0].detach().cpu().numpy().astype(int).tolist()
        log["leg_mask"].append(leg_mask_vals)
        log["leg_fault_active"].append(leg_act_vals)

        if cur_wp >= len(waypoints):
            # ゴール到達で任意に早期終了したければ break 可能
            pass

    # メタ情報
    log["meta"] = dict(
        waypoints=waypoints,
        wp_radius=args.wp_radius,
        push_times=push_times,
        push_force=push_force,
        episode_sec=args.episode_sec,
        dt=dt
    )

    with open(args.outfile, "w") as f:
        json.dump(log, f, indent=2)
    print(f"[DONE] Saved log to {args.outfile}  steps={len(log['t'])}  reached_wp={cur_wp}/{len(waypoints)}")


if __name__ == "__main__":
    main()
```

```python
%%writefile render_mesh_hud.py
import argparse, json, math, os, numpy as np, torch, imageio
import matplotlib.pyplot as plt
import genesis as gs
import sys
sys.path.append("examples/locomotion")
from go2_env import Go2Env
from go2_train import get_cfgs

# カラー定義
COL_NORMAL = (1.0,1.0,1.0)
COL_FAULT  = (1.0,0.85,0.0)
COL_RECOV1 = (1.0,0.6,0.0)
COL_RECOV2 = (0.6,1.0,0.2)

def lerp(a,b,t): return tuple(a[i]*(1-t)+b[i]*t for i in range(3))

def compute_leg_color(mask_val, active_flag):
    # mask_val: 0.4~1.0 付近
    if active_flag:
        # まだ回復中
        # 低いと黄色→オレンジ→緑に遷移
        if mask_val < 0.7:
            # fault zone
            return COL_FAULT
        elif mask_val < 0.9:
            # recovering
            u = (mask_val-0.7)/(0.2)
            return lerp(COL_RECOV1, COL_RECOV2, max(0,min(1,u)))
        else:
            return COL_RECOV2
    else:
        # 完全回復
        return COL_NORMAL

def draw_hud_overlay(img, t, meta, wp_index, total_wp, fault_event, push_flag):
    # 簡易 HUD を matplotlib で合成
    h, w, _ = img.shape
    fig = plt.figure(figsize=(w/100, h/100), dpi=100)
    ax = fig.add_axes([0,0,1,1])
    ax.imshow(img)
    ax.axis("off")

    # バー
    prog = wp_index/total_wp
    ax.text(10,20,f"t={t:5.1f}s   WP {wp_index}/{total_wp}", color="white",
            fontsize=12, fontweight="bold", bbox=dict(facecolor="black", alpha=0.4, pad=4))
    # Progress bar
    ax.add_patch(plt.Rectangle((10,40), 200, 12, color="white", alpha=0.2))
    ax.add_patch(plt.Rectangle((10,40), 200*prog,12, color="limegreen", alpha=0.9))

    x_off = 220
    if fault_event:
        ax.text(x_off, 20, "FAULT!", color="black",
                bbox=dict(facecolor="gold", alpha=0.9, pad=4), fontsize=12)
        x_off += 90
    if push_flag:
        ax.text(x_off, 20, "PUSH", color="white",
                bbox=dict(facecolor="red", alpha=0.7, pad=4), fontsize=12)
    if wp_index==total_wp:
        ax.text(w/2, h*0.08, "MISSION COMPLETE", color="white",
                fontsize=18, ha='center', bbox=dict(facecolor="green", alpha=0.7, pad=6))

    fig.canvas.draw()
    buf = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
    out = buf.reshape(h, w, 3)
    plt.close(fig)
    return out

def main():
    ap = argparse.ArgumentParser()
    ap.add_argument("--logfile", type=str, default="log_baseline.json")
    ap.add_argument("--outfile", type=str, default="demo_mesh.mp4")
    ap.add_argument("--fps", type=int, default=30)
    ap.add_argument("--cam-dist", type=float, default=2.5)
    ap.add_argument("--cam-height", type=float, default=1.2)
    ap.add_argument("--cam_pitch_deg", type=float, default=25)
    ap.add_argument("--width", type=int, default=640)
    ap.add_argument("--height", type=int, default=480)
    ap.add_argument("--random-action-std", type=float, default=0.05)
    args = ap.parse_args()

    with open(args.logfile) as f:
        log = json.load(f)

    meta = log["meta"]
    waypoints = meta["waypoints"]
    n_wp = len(waypoints)
    dt = meta["dt"]

    T = log["t"]
    num_steps = len(T)

    gs.init(logging_level="warning")
    env_cfg, obs_cfg, reward_cfg, command_cfg = get_cfgs()
    # 評価用: 短め・1 env
    env_cfg["episode_length_s"] = meta["episode_sec"]
    base = Go2Env(
        num_envs=1,
        env_cfg=env_cfg,
        obs_cfg=obs_cfg,
        reward_cfg=reward_cfg,
        command_cfg=command_cfg,
        show_viewer=False
    )

    # カメラをシーンに追加
    scene = base.scene
    cam = scene.add_camera(
        width=args.width,
        height=args.height,
        fov=55
    )

    # 色変更のための関節/リンク参照
    # 簡易的には各脚の大腿リンク名前でマテリアルアクセス（URDFに応じて調整必要）
    # ここでは例としてリンク名パターンを配列
    leg_link_candidates = [
        ["FR_thigh_link","FR_calf_link"],
        ["FL_thigh_link","FL_calf_link"],
        ["RR_thigh_link","RR_calf_link"],
        ["RL_thigh_link","RL_calf_link"],
    ]
    link_handles = []
    for group in leg_link_candidates:
        sub = []
        for name in group:
            try:
                sub.append(base.robot.get_link(name))
            except:
                pass
        link_handles.append(sub)

    writer = imageio.get_writer(args.outfile, fps=args.fps)

    # 再生ステップをログの dt からレンダ fps に合わせて間引き/補間
    step_interval = max(1, int(round((1/args.fps)/dt)))

    print(f"[Render] steps={num_steps} interval={step_interval} -> frames≈{num_steps/step_interval}")

    # 簡易実行ループ
    for idx in range(0, num_steps, step_interval):
        # シミュを dt*step_interval 進める（ランダム安定姿勢ぽく）
        for _ in range(step_interval):
            act = torch.randn((1, base.num_actions))*args.random_action_std
            base.step(act.numpy() if isinstance(act, torch.Tensor) else act)

        # ベース位置
        bp = base.base_pos[0].detach().cpu().numpy()

        # ログに基づき脚カラー設定
        leg_mask = log["leg_mask"][idx]      # list of 4
        leg_active = log["leg_fault_active"][idx]
        for leg_id, (mask_val, active_flag) in enumerate(zip(leg_mask, leg_active)):
            color = compute_leg_color(mask_val, active_flag==1)
            for lh in link_handles[leg_id]:
                try:
                    lh.set_color(color)  # Genesis のリンクが set_color サポートの場合
                except:
                    pass

        # カメラ位置計算（追従: 後ろ上方）
        yaw = 0.0  # とりあえず x 方向進行前提
        dist = args.cam_dist
        cam_x = bp[0] - dist*math.cos(yaw)
        cam_y = bp[1] - dist*math.sin(yaw)
        cam_z = bp[2] + args.cam_height
        lookat = (bp[0], bp[1], bp[2]+0.2)
        cam.set_pose(position=(cam_x, cam_y, cam_z), look_at=lookat)

        # レンダ
        scene.step()  # 1ステップ済後の状態
        rgb = cam.render()  # (H,W,3) uint8

        hud_frame = draw_hud_overlay(
            rgb,
            T[idx],
            meta,
            wp_index=log["wp_index"][idx],
            total_wp=n_wp,
            fault_event=bool(log["fault_event"][idx]),
            push_flag=bool(log["push_flag"][idx])
        )

        writer.append_data(hud_frame)

    writer.close()
    print(f"[DONE] Saved video: {args.outfile}")

if __name__ == "__main__":
    main()
```
